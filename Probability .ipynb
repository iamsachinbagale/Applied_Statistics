```python
import numpy as np
import pandas as pd
from scipy.special import ndtri  #Used to calculate theoretical z values.
import matplotlib.pyplot as plt

data_points = np.random.uniform(0,1,50)
# print(data_points)
df=pd.DataFrame(data_points,columns=['Points'])
# print(df)
df=df.sort_values(by=['Points'],ascending=True).reset_index(drop=True)
# df
df['count']=df.index+1
# df
n_rows=df.shape[0]
# n_rows
df['percentile_area']=(df['count'])/n_rows
# df
df['z_theortical']=ndtri(df['percentile_area'])
# df
df['z_actual']=(df['Points']-df['Points'].mean())/df['Points'].std()
# df
plt.scatter(df['z_theortical'],df['z_actual'])
plt.plot([-2,-1,0,1,2],[-2,-1,0,1,2]) 
plt.show()
```

H1: The average sales differ in size for all three people:  1  2  3 


N = 14
T = 1500 + 1600 + 2500 = 5600
Correlation factor = T2 / N = 56002 / 14 = 31,360,000/14 = 2240000

SST =x12 + x22 + x32 - T2/N = 2390000 - 2240000
= 150000

SSB = (x1)2/ n1 + (x2)2/ n2 + (x3)2/ n3 - T2 /N
= 450000 + 640000 + 1250000 - 2240000
= 2340000 - 2240000 = 100000

SSW= 150000 - 100000 = 50000

Degrees of freedom :
r-1 = 5-1 = 4
N-r = 14 - 5 = 9
N-1 = 14 -1 = 13

MSB = SSB / r-1 = 100000/4 = 25000
MSW = SSW / n-r = 50000 / 9 = 5555.56

MSB > MSW
Fstat = MSB / MSW = 25000/5555.56 = 4.499
Degree = (4,9) at 5%
Fcritical = 3.63
Fstat > Fcritical

Hence, the null hypothesis is rejected.
